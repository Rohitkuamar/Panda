# Pandas Series

A Series is a one-dimensional labeled array capable of holding any data type. The axis labels are collectively called the index. Allows us to analyze big data and make conclusions based on statistical theories. Clean messy data sets, and make them readable and relevant.

## Creating Series
There are multiple ways to create a pandas Series:

```python

import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c', 'd']
arr_label = np.array(labels)
arr_label

```

```python
my_list = [10, 20, 30, 40]
arr = np.array(my_list)
arr
```

```python
pd.Series(arr)
```

```python
d = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
pd.Series(data=d)
```

```python
labels
my_list
pd.Series(data=my_list, index=labels)
```

```python
pd.Series(data=arr)
pd.Series(data=d)
pd.Series(data=d, index=['a', 'b', 'c'])
pd.Series(data=[sum, print, len])


```

```python
ser1 = pd.Series(data=[1, 2, 3, 4], index=['USA', 'Germany', 'USSR', 'Japan'])
ser2 = pd.Series(data=[1, 2, 5, 4], index=['USA', 'Germany', 'Italy', 'Japan'])
ser1
ser2

```

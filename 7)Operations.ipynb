**DataFrames Basic Operations**

```python
df1 = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
})
```

```python
df1.shape # Get the shape of the DataFrame (number of rows and columns)
```

```python
df1.columns # Get the column names of the DataFrame
```

```python
df1.info() # Get information about the DataFrame, including data types and memory usage
```

```python
df1.describe() # Get summary statistics for the DataFrame
```

```python
df1['A'] + 10 # Add 10 to each element in column 'A'
```

DataFrames Applying Functions

```python
def square(x):
    return x ** 2

df1['A'].apply(square) # Apply the square function to each element in column 'A'    
```

```python
df1['D'] = df1['A'].apply(square)  # Create a new column 'D' by applying the square function to column 'A'
```

```python
df1
```

```python
df1['D'] = df1['B'].apply(lambda x: x * 2)  # Create a new column 'D' by applying a lambda function to column 'B'
```

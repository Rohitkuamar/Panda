```python
import numpy as np 
import pandas as pd 
```

**Finding Missing Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum() # number of missing values in each column
```

```python
df.isna().any() # any missing values in each column
```

Removing Missing Data

```python
df.dropna() # drop rows with any missing values
```

```python
df
```

```python
df.dropna(thresh=1) # drop rows with less than 3 non-missing values
```

Filling the missing Data

```python
df
```

```python
df.fillna(0) # fill missing values with 0
```

```python
values = {'A':0,'B':100,"C":300,'D':400}  # fill missing values with different values for each column
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean()) # fill missing values with the mean of each column (only for numeric columns)
```
